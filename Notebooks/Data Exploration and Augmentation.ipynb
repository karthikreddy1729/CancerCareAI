### **Data Augmentation**

```python
import pandas as pd
data = pd.read_csv("/lakehouse/default/Files/Immunotherapy.csv")
```

```python
data.drop('Time', axis=1)
```

```python
import ctgan
```

```python
model = ctgan.CTGAN()  
```

```python
model.fit(data)
```

```python
number_of_samples = 250
synthetic_data = model.sample(number_of_samples)
```

```python
synthetic_data
```

```python
import pandas as pd

new_data_250 = synthetic_data[(synthetic_data >= 0).all(axis=1)]
```

```python
new_data_250.to_csv('/lakehouse/default/Files/Immunotherapy_.csv')
```
